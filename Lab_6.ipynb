### Define and generate data

```r
n = 50
mean_1 = 11
sd_1 = 1

mean_2 = 12
sd_2 = 0.5

y1 = rnorm(n,mean_1,sd_1)
y2 = rnorm(n,mean_2,sd_2)

# move data to df
data_df <- data.frame(co1 = y1, col2 = y2)
colnames(data_df) <- c("y1","y2")
```

```r
# create data for stan models

stan_model_data = list(n=n, y1=y1, y2=y2)
```

### Fit the model

```r
model2_samples <- stan(file = 'model2.stan', data = stan_model_data)
```

```r
# move results to dataframe

model2_samples_df = as.data.frame(model2_samples)
```

### Plot the results

```r
library(bayesplot)

color_scheme_set("red")
options(repr.plot.width=15, repr.plot.height=8)
```

#### Intervals

```r
mcmc_intervals(model2_samples_df, pars = c("mu1", "mu2","sigma")) + theme(text = element_text(size = 20))
```

#### Posterior distributions of parameters

```r
# inspect the real values of y1 and y2 means

summary(y1)
summary(y2)
```

```r
# inspect the histograms of y1 and y2

mcmc_hist(data_df, pars = c("y1", "y2"), bins=50) + 
    ylim(c(0, 10)) + 
    xlim(c(8,14)) +
    theme(text = element_text(size = 20))  + 
    yaxis_ticks(on = TRUE) + 
    yaxis_text(on = TRUE)
```

```r
# plot the histograms of posterior distributions of parameters

mcmc_hist(model2_samples_df, pars = c("mu1", "mu2", "sigma")) + theme(text = element_text(size = 20)) 
```

As the 'real' mean of y1 is 11.15  it is not surprising that the posterior distribution of y1 is shifted to the right

#### Density plot of posterior distribution of means y1 and y2 - another approach to visualisation

```r
options(repr.plot.width=10, repr.plot.height=10)

posterior_2 <- as.matrix(model2_samples)

plot_title <- ggtitle("Posterior distributions",
                      "with medians and 95% intervals")
mcmc_areas(posterior_2,
           pars = c("mu1", "mu2"),
           prob = 0.95) + plot_title
```

### Credible intervals

#### Credible intervals using quantiles with quantile function

```r
print("MEAN 1")
quantile(model2_samples_df$mu1, probs = c(0.025, 0.975))
print("MEAN 2")
quantile(model2_samples_df$mu2, probs = c(0.025, 0.975))
print("SIGMA")
quantile(model2_samples_df$sigma, probs = c(0.025, 0.975))
```

#### Credible intervals using summary function

```r
summary(model2_samples, probs = c(0.025, 0.975))$summary
```

### Samples from the posterior distribution of the DIFFERENCE between the means

```r
# calculate difference
diff_mu1_mu2 = model2_samples_df$mu2 - model2_samples_df$mu1

# and insert to df
model2_samples_df$mean_diff = diff_mu1_mu2
```

#### Histograms of posterior distributions including the difference between means

```r
mcmc_hist(model2_samples_df, pars = c("mu1", "mu2", "sigma", "mean_diff")) + theme(text = element_text(size = 20)) 
```

#### 95% Credible intervals using quantiles with quantile function

```r
cat("mean of diff in means: ", mean(model2_samples_df$mean_diff))
quantile(model2_samples_df$mean_diff, c(0.025, 0.975))
```

#### 95% Confidence intervals (Frequentist analysis)

```r
mean_diff = y2 - y1

t.test(mean_diff)
```

In this case credible and confidence intervals are similar

### Model 3: with two means and two sigmas

#### Define and generate data

```r
n = 10000
mean_1 = 11
sd_1 = 1

mean_2 = 12
sd_2 = 0.5

y1 = rnorm(n,mean_1,sd_1)
y2 = rnorm(n,mean_2,sd_2)

data_df <- data.frame(co1 = y1, col2 = y2)
colnames(data_df) <- c("y1","y2")
```

```r
# create data for stan models

stan_model_data = list(n=n, y1=y1, y2=y2)
```

```r
# inspect distribution of created data: y1 and y2 with density plots

dens_plot = mcmc_dens(data_df, pars = c("y1", "y2")) + 
    ylim(c(0, 1)) +
    xlim(c(8,14)) +
    theme(text = element_text(size = 15))+ 
    yaxis_ticks(on = TRUE) + 
    yaxis_text(on = TRUE)

dens_plot
```

#### Fit model 3

```r
# fit the model
model3_samples <- stan(file = 'model3.stan', data = stan_model_data)
```

#### Compare model 2 (one sigma) and model 3 (two sigmas)

```r
# fit again model 2
model2_samples <- stan(file = 'model2.stan', data = stan_model_data)
```

```r
# inspect the real mean of y1 and y2

summary(y1)
summary(y2)
```

```r
# compare credible intervals of model 2 (one sigma) and model 3 (two sigmas)

summary(model2_samples, probs = c(0.025, 0.975))$summary
summary(model3_samples, probs = c(0.025, 0.975))$summary
```

- Model 3 better estimate y1 and y2 means (but not much)
- For y1 the model 3 has much wider credible intervals, for y2 narrower - this (maybe) shows the differentiation of the distribution of means y1 and y2, when two sigmas allowed

#### Compare visually posterior distribution of means

```r
posterior_2 <- as.matrix(model2_samples)

plot_title <- ggtitle("Posterior distributions od model 2",
                      "with medians and 95% intervals")
posterior_2_plot = mcmc_areas(posterior_2,
           pars = c("mu1", "mu2"),
           prob = 0.95) +
    plot_title + 
    theme(text = element_text(size = 15))
```

```r
posterior_3 <- as.matrix(model3_samples)

plot_title <- ggtitle("Posterior distributions of model 3",
                      "with medians and 95% intervals")

posterior_3_plot = mcmc_areas(posterior_3,
           pars = c("mu1", "mu2"),
           prob = 0.95) + 
    plot_title + 
    theme(text = element_text(size = 15))
```

```r
bayesplot_grid(posterior_2_plot, posterior_3_plot, dens_plot)
```

Note that posterior distribution of means from models with two sigmas are much similar to orginal distribution of data

### Model 4: with the student likelihood instead normal likelihood

#### Define data and fit model

```r
# define degrees of freedom for student distribution
student_dof = 4

# define data for stan models
model4_data = list(n=n, y1=y1, y2=y2, nu=student_dof)

# fit model
model4_samples <- stan(file = 'model4.stan', data = model4_data)
```

#### Compare model 3 (normal likelihood) and model 4 (student likelihood)

```r
# inspect the real mean o y1 and y2

summary(y1)
summary(y2)
```

```r
# compare credible intervals of model 3 (normal dist) and model 4(student dist)

summary(model3_samples, probs = c(0.025, 0.975))$summary
summary(model4_samples, probs = c(0.025, 0.975))$summary
```

Model 3 is sometimes better than Model 4, but it strongly depends on the generated data (y1, y2). 

Because we have very small dataset - 50 elements, if there are a lot of extreme values (empirical mean of y1 and y2 is shifted from desired), Model 4 will treat those values as noise. In different runs (when generated different data), when data was more balanced (mean close to 11 in y1 and 12 in y2), Model 4 was sometimes better than Model 3.

#### Compare visually posterior distributions of means

```r
options(repr.plot.width=10, repr.plot.height=10)
```

```r
posterior_3 <- as.matrix(model3_samples)
plot_title <- ggtitle("Posterior distributions of model 3",
                      "with medians and 95% intervals")
posterior_3_plot = mcmc_areas(posterior_3,
           pars = c("mu1", "mu2"),
           prob = 0.95) + plot_title
```

```r
posterior_4 <- as.matrix(model4_samples)
plot_title <- ggtitle("Posterior distributions of model 4",
                      "with medians and 95% intervals")
posterior_4_plot = mcmc_areas(posterior_4,
           pars = c("mu1", "mu2"),
           prob = 0.95) + plot_title
```

```r
bayesplot_grid(posterior_3_plot,posterior_4_plot) 
```

Distributions are very similar

## Testing normal and student models in the case of outliers

#### Create outlier - has to be stronger than 10, as y1 already contains values close to 10 ;)

```r
# copy y1 to y1_o
y1_o = y1

# change value of 50. element of y1_o list on 20
y1_o[50] = 20
```

```r
# inspect the current means of y1 and y1_o

summary(y1)
summary(y1_o)
```

It is visible, that oulier strongly affect the mean of the data

```r
# fit again model 3

stan_model_data_outlier = list(n=n, y1=y1_o, y2=y2)
model3_o_samples <- stan(file = 'model3.stan', data = stan_model_data_outlier)
```

```r
# fit again model 4

student_dof = 4

stan_model4_data_outlier = list(n=n, y1=y1_o, y2=y2, nu=student_dof)
model4_o_samples <- stan(file = 'model4.stan', data = stan_model4_data_outlier)
```

#### Compare model 3 (normal likelihood) with outlier and model 4 (student likelihood) with outlier

```r
# summary of the models

summary(model3_o_samples, probs = c(0.025, 0.975))$summary
summary(model4_o_samples, probs = c(0.025, 0.975))$summary
```

Note that model mostly 4 ignored the influence of outlier and its estimates of mean y1 remained almost the same, while model 3 incorporated this information and changed its estimates.

#### Compare visually posterior distributions of means

```r
posterior_3_o <- as.matrix(model3_o_samples)

plot_title <- ggtitle("Posterior distributions of model 3 with oulier",
                      "with medians and 95% intervals")
posterior_3_o_plot = mcmc_areas(posterior_3_o,
           pars = c("mu1", "mu2"),
           prob = 0.95) + plot_title
```

```r
posterior_4_o <- as.matrix(model4_o_samples)

plot_title <- ggtitle("Posterior distributions of model 4 with oulier",
                      "with medians and 95% intervals")
posterior_4_o_plot = mcmc_areas(posterior_4_o,
           pars = c("mu1", "mu2"),
           prob = 0.95) + plot_title
```

```r
bayesplot_grid(posterior_3_o_plot,posterior_4_o_plot, posterior_3_plot, posterior_4_plot) 
```

### Conclusions

- Model with two sigmas is more robust - obvious
- Model with student likelihood is similar to the one with normal likelihood in balanced data
- Model with student likelihood is much better (stable) in case of outliers
